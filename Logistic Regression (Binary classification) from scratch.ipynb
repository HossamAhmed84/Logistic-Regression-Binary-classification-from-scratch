## Task2: Define the Logistic Sigmid function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

## Task2: Define the y_predict function

```python
def hypothesis(X, theta):
    z = np.dot(theta, X.T)
    return 1/(1+np.exp(-(z))) - 0.0000001
```

Comment:  deducted 0.0000001 from the output here because of this expression because If the outcome of the hypothesis expression comes out to be 1, then this expression will turn out to be the log of zero

## Task3: Define the Cost function

```python
def cost(X, y, theta):
    y1 = hypothesis(X, theta)
    return -(1/len(X)) * np.sum(y*np.log(y1) + (1-y)*np.log(1-y1))
```

## Task4: Define gradient descent and update the theta values function 

```python
def gradient_descent(X, y, theta, alpha, epochs):
    m =len(X)
    J = [cost(X, y, theta)] 
    for i in range(0, epochs):
        h = hypothesis(X, theta)
        for i in range(0, len(X.columns)):
            theta[i] -= (alpha/m) * np.sum((h-y)*X.iloc[:, i])
        J.append(cost(X, y, theta))
    return J, theta
```

## Task5: Define final predict values function and accuracy

### It will be add variables called epochs which define the number of itrations you will run this function
### alpha is the value will be extract eachtime to get new gradient descent normally it can be start by 0.0001 to make sure not miss the point of minmum cost
